## What is it?
***
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python programming language [2].<br>
The library utilises a tree-like hierarchy structure to store its objects for each plot [9]. Matplotlib is a free open-sourced library<br>
which is continually improving. 

<br>

## Brief History
***
Matplotlib development started in the early 2000’s by John D. Hunter, a neurobiologist that emulated MATLAB’s capabilities to help<br> 
visualize electronic signals in the brain of epilepsy patients [12]. Despite MATLAB’s limitations, it’s approach to plotting was retained<br>
in the pyplot sub module which was developed to interface with Matplotlib through a collection of functions [2]. 

<br>

## Pyplot
### Overview
***
This sub module is typically imported from the Matplotlib library under the “plt”alias name and can carry out the majority of Matplotlib’s<br> 
plotting capabilities with relatively intuitive syntax. Plotting via pyplot is considered the stateful approach to plotting with Matplotlib,<br>
where calls are made via plt.plot() and other top level pyplot functions which preserve states to keep track of the current figure and axes [9].<br>
As described in RealPython, the simplicity of pyplot’s plotting syntax can be attributed to its wrapper functions that essentially call <br>
Matplotlib’s more complicated object orientated interface. Image 1 below details the object hierarchy in Matplotlib, where the Figure is<br>
the outermost object and holds at least one axes and each axes holds at least 2 axis depending on the dimension of the given plot.

### Plotting
***

The pyplot library can do many different plots via different functions calls. The arguments accepted by the different pyplot plots are<br>

#### Breakdown of plt.plot()
[Documentation:](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)  _matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)_

This *plot* functon:
- Accepts list of arguements (**args*) which can form a 2d-array via:
  - Lists of equal lengths (as used in *line_plot* function in next code cell).
  - Numpy arrays of equal length.
  - Keyword strings, such as  reference to pandas dataframe columns.
-  View limits autoscale to data limits by default as scalex and scaley are set to True. Can be altered in the function or after the fact.
-  Data input is defaulted to None but one can pass data object if the object variables are declared.
-  Keyword arguements (***kwargs*) declared in or passed into the *plt.plot()* can format properties  like linewidth and marker color. Also <br>
possible to shortcut string notation for some formatting operations (as used in *line_plot* function below, e.g blue dashes line = 'b--').


````python
# Import the necessary libraries
%matplotlib inline
import matplotlib.pyplot as plt            # for plotting
import numpy as np                         # for numerical operations
import pandas as pd                        # for making dataframes
import requests as rq                      # for making requests from websites
from sklearn import datasets               # for importing iris dataset
import scipy.stats

# default figure sizes of future plots in the notebook
plt.rcParams['figure.figsize'] = [10, 6]      
````

````python

# define plot variables
x = np.arange(0, 6, 0.5)                      # generate array of 0-5 in increments of 0.5
y1 = x ** 2
y2 = x ** 3
y3 = x ** 4

# define plot labels
xy1_label = 'f(x)=x'
xy2_label = 'g(x)=x^2'
xy3_label = 'h(x)=x^3'

def line_plot():  
# pass the following arguements into plot function():
    # x and y lists of equal length
    # Shorthand string notation for colour / linestyle
    # linewidth value & label string (later used to format legend)
    plt.plot(x, y1, 'b--', linewidth = 3, label = xy1_label)  
    plt.plot(x, y2, 'g--', linewidth = 3, label = xy2_label) 
    
line_plot()

# Add Figure Reference
plt.figtext(0.5, 0.0, 'Figure 1', wrap = True, horizontalalignment = 'center', fontsize = 15);
````

### Additional Formatting 
***

````python
# Additional plot formatting (declared outside of the lineplot function)

plt.clf() # clear current figure

def format_plot():
    '''This function directs formatting functions to current axes via pyplot module'''
    
    # plot format variables
    x_axis_label = 'x_axis'
    y_axis_label = 'y_axis'
    default_fontsize = 12
    label_pad = 5
    axis_font = {'family': 'serif',
        'color':  'red',
        'weight': 'bold',
        'size': 15,
        }

    # set axis labels, declare fontsize and distance from axis (labelpad)
    plt.xlabel(x_axis_label, fontdict = axis_font, labelpad = label_pad)
    plt.ylabel(y_axis_label, fontdict = axis_font, labelpad = label_pad)
    
    # alter scaling of y axis (this can be done in plot function either)
    plt.ylim(top = 200)
    
    # set fontsize of axis tick labels
    plt.xticks(fontsize = default_fontsize)
    plt.yticks(fontsize = default_fontsize)
    
    # alter tick frequency of x-axis to 0.5
    plt.xticks(np.arange(min(x), max(x) + 1, 0.5))
    
    # set current axis to a variable to call functions on it
    ax = plt.gca()                  # define ax variable as current axes if it exists
    ax.set_facecolor('silver')      # call set_facecolur method current axes
    ax.grid()                       # apply grid to current axes
    
    # annotate datapoint labels on data series
    # loop over index & val of series, then add text via annotate function
    # Adopted from: https://queirozf.com/entries/add-labels-and-text-to-matplotlib-plots-annotation-examples
    for index, value in enumerate(y2):
        value = value.round(1)
        ax.annotate(value, (x[index], y2[index]), textcoords = "offset points", xytext = (-10, 5), 
                    ha = 'center', color = 'darkgreen')
    
    # Place a legend on the current Axes
    # (labels taken from artist as label declared in plot function)
    # Note: labels can also be passed as a list to plt.legend() 
        # bbox_to_anchor and loc are used together to position legend
        # mode set to expand stretches legend accross axes length
        # borderaxespad control distance between legend and legend border
    ax.legend(bbox_to_anchor = (0, 1, 1, 0.2), loc = "lower left", mode = "expand",
               borderaxespad = 0.5, ncol = 3, fontsize = 15)

line_plot()      # plot 2 series
format_plot()    # additional formatting of plot
plt.figtext(0.5, -0.05, 'Figure 2', wrap = True, horizontalalignment = 'center', fontsize = 15);
````

### Preserving States
***
When looking at Figure 2 above, the *lineplot()* function plotted two series via two consecutive *plot()* calls. This is an example of how the<br>
pyplot module can make it easier to plot. Pyplot can preserve various states enabling it to keep track of the current figure, whereby functions<br>
called are directed to the current axes. Consecutive *plt.plot()* calls are essentially directing each plot to the same axis. This behaviour can<br>
be explained when one looks at whats happening under the hood.

````python
# Reference [9]
# Abridged version of plt.plot()

def plot(*args, **kwargs):  
    ax = plt.gca()                      # Get the current Axes instance of the current figure or create one if doesnt exist.
    return ax.plot(*args, **kwargs)

def gca(**kwargs):                                 
    return plt.gcf().gca(**kwargs)      # Get the current Axes of the current Figure.
````


<br>

The same logic can be used when explaining how calling the *format_plot()* function decorated the plot shown in Figure 2 despite the absence <br>
of *plot()* in the format_plot() function. This is because the formatting function calls (e.g plt.xticks()) were directed to the current axes. RealPython<br>
(2018, Stateful Versus Stateless Approaches) explains how almost all pyplot function implicitily refer to an existing Figure and current Axes, <br>
or create a new one if none exist.

### Generating subplots:
#### The stateful approach: plt.subplot()
***
As per the offical [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html) the *subplot* funcion adds an axes to the current figure or retrieves the current axis if it exists. The position<br>
of the subplot is determined by the following integer arguements (nrows, ncols, index), whereby a 2x2 grid will be declared via *subplot(2, 2, n)*,<br>where n is the index of the plot.


````python
# generate new figure, or activate existing
# (although subplot generates new figure this was used to control Fig size)
fig = plt.figure(figsize = (12, 8))
plt.figtext(0.5, -0.05, 'Figure 3', wrap = True, horizontalalignment = 'center', fontsize = 15);

# plot 1: axes index position 1 in 2x2 grid
plt.subplot(2, 2, 1)   
plt.title('Subplot 1 (histogram plot)')
rng = np.random.default_rng()                                       # instansiate object of numpy generator class
normal_array = rng.normal(0, 1, 100000)                             # get normal standard normal distribution
plt.hist(normal_array, bins = 50)       


# plot 2: axes index position 2 of 2x2 grid
plt.subplot(2, 2, 2)
plt.title('Subplot 2 (bar plot)')     
plt.bar(['label1', 'label2', 'label3', 'label4'], [1, 2, 3, 4])


# plot 3: axes index position 3 of 2x2 grid
plt.subplot(2, 2, 3) 
plt.title('Subplot 3 (scatter plot)')
 
# scatter plot data adopted from:
# https://numpy.org/devdocs/reference/random/generated/numpy.random.Generator.multivariate_normal.html
mean = [0, 0]
covariance_matrix = [[1, 0.95], [0.95, 1]]
scatter_data = rng.multivariate_normal(mean, covariance_matrix, 100)  # generate two correlated normal arrays
plt.scatter(scatter_data[:, 0], scatter_data[:, 1])                   # plot col 1 vs. col 2 of 2d array


# plot 4: index position 4 of 2x2 grid
plt.subplot(2, 2, 4)
line_plot()                   
format_plot()
plt.tight_layout(pad = 2)

string_spacer = [' ' * 11 for i in range(4)]
print('\n{}Here we can see adding additional axes to a figure can be useful for displaying more plots in a grid\n\
{}format. Notice again how the preservation of states is at play as only the last subplot was decorated\n\
{}via the format_plot() call. The object orientated (stateless approach) allows easier access to\n\
{}subplots and will be briefly explored to show its benifit.\n'.format(*string_spacer));   
````

#### The statefless approach: plt.subplots()

The subplots() function generates a figure similar to subplot(), however the subplots figure can hold an array of axes objects. As described by <br>
Nyquist (2020), these axes object can be retrieved and plotting agruements called on them. This differs to *subplot*  generated figures which typically<br>
use pyplot functions to direct plotting aruements to the current axes. See Figure 4 below for arguably cleaner approach with a higher degree of<br>
control.

````python
fig, axes = plt.subplots(2, 2, figsize = (12, 8))
plt.figtext(0.5, -0.05, 'Figure 4', wrap = True, horizontalalignment = 'center', fontsize = 15);

# index matrix position and call relevant plot functions
axes[0, 0].hist(normal_array, bins = 50)
axes[0, 1].bar(['label1', 'label2', 'label3', 'label4'], [1, 2, 3, 4])
axes[1, 0].scatter(scatter_data[:, 0], scatter_data[:, 1])
axes[1, 1].plot(x, y1, 'b--', linewidth = 3) 
axes[1, 1].plot(x, y2, 'g--', linewidth = 3) 

# loop over axes indices and format given axes
# leverage enumerate function to label axes titles
for count, ax in enumerate(fig.axes):  
    ax.set_facecolor('silver')
    ax.set_title(str("Subplot ") + str(count + 1))     
    ax.set_xlabel('x_axis')
    ax.set_ylabel('y_axis')
    
fig.tight_layout(pad = 2.5)


string_spacer = [' ' * 11 for i in range(4)]
print('\n{}With the object orientated approach we can directly reference any axes by indexing its \n\
{}matrix position in the grid. This approach gives a greater degree of control and is\n\
{}generally a tidier approach.'.format(*string_spacer));
````

### Histogram

 #### Plot Description
 ***
The histogram plot (or variations of it) is one of the most fundamental plots used in the arena of data science. This 2D plot is used to view<br>
the distribution of continuous (floats) or discrete (integer) data on an interval scale where the variables have order. A histogram plot consists<br>
of frequency or count on the y-axis and the number line pertaining to the variables on the x-axis. The x-axis is particularly interesting as the<br>
data is grouped into numerical intervals called bins. Weitz (2020) details how the frequency of each bin is represented by its area, whereby<br>
the area of all the bins represents a probability of one. In other words one can get frequency of each bin by taking the area under its bar <br>(i.e: bin height
multiply by bin width). He then goes on to explain how numbers that exactly equal the bin bounderies will be assigned to the <br>
left or right of the given interval depending on the software used to generate the plot.

Lets explore the histogram using Matplotlib. One can adjust many of the features of a histogram by passing keyword arguements to the *hist()*<br>
function of the pyplot module (see [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)). One can adjust these parameters to their needs or alternatively not declare them and therefore take<br>their 
default values. For example the code used to generate Figure 5 below does not specify a density parameter which then defaults the plot to <br>
displaying count on the y-axis. Subplot 1 and 2 of figure 5 shows the difference in applying a different number of bins to the same array,<br>
where subplot 1 looks essentially looks like a barchart as the number of bins is set to a high value, whereas subplot 2 shows the same data<br>
grouped into 5 bins (see code print statement for more information on the grouping).


````python
# exploring how bins work (using discrete data)

rng = np.random.default_rng(100)                              # instansiate seeded object of numpy generator class
rand_ints = rng.integers(0, 11, 100)                          # generate 100 random ints from 0 to 10   

# generate new figure, or activate existing
fig = plt.figure(figsize = (15, 5))
plt.figtext(0.5, -0.05, 'Figure 5', wrap = True, horizontalalignment = 'center', fontsize = 15);
                        
# plot 1: large number of bins
plt.subplot(121)                                              # create figure, 1row, 2 cols, index 1            
plt.title('Subplot 1 (bins = 100)')                           # title on subplot
plt.ylabel('count')                                           # label y axis as count as density = False
plt.hist(rand_ints, bins = 100)                               # plot with high bin number

# plot 2: medium number of bins
plt.subplot(122)                                              # add to  figure at index 2            
plt.title('Subplot 2 (bins = 5)')                             # title on subplot
plt.ylabel('count')                                           # label y axis as count as density = False
plt.hist(rand_ints, bins = 5, rwidth = 0.9)                   # plot with low bin num and seperate bars for comparison

unique, counts = np.unique(rand_ints, return_counts = True)                # return unique values and their counts 
print('{}x-axis{}\n{}y-axis{}\n'.format(' '*40,unique, ' '*40, counts))    # print x-axis and y-axis arrays

print('  Subplot 1 values are not grouped as bin count is high, \
subplot 2 has grouped all subplot 1 data (11 bars) accross 5 bins.\n\
  To account for the odd number in the grouping, subplot 1 values\
 for index position 8, 9 and 10 are grouped into the\n\
  final bin (8 - 10 range) of subplot 2, value = 13 + 9 + 7 = 29.\
 One should chose a bin number from approx 5 to 20\n\
  depending on sample size, and where possible select\
 one where where your sample-size is divisible by the bin number.\n');
````

#### Modelling Real World Data

Histograms are mostly used to model probability distributions (i.e: where frequency distribution is the y-axis label). When modelled in <br>
this way the area of the bins correspond to the frequency of the interval and therefore the probability that values lie in that range. This<br>
is somewhat different to Figure 5 where the height of the bin corresponded to a count value. When measuring frequency, histograms can<br>
model popular probability distributions that are seen in the real world. The ability to model real world scenarious allows us to predict<br>
the probability of certain events occuring, which is of inestimable benifit. Onnen (2021) describes how data scientists can join the dots <br>
of the 3-point-estimation technique to avoid business pitfalls, i.e: model the best case, likely case and worst case to help describe possible<br>
outcomes with their associated probabilities. Lets now explore how a histogram can be useful in modelling a real world scenario that is<br>
described by a normal distribution. 

**Scenario:**<br>
Model Men's height, with a normal distirbution of mean 70 inches and of standard deviation 3 inches.<br>
Note: Brooks-Bartlett (2018) details how a given Probability Density Function (PDF) can be used to calculate probabilites over a range<br>
by integrating the area under the curve. This will be leveraged in Figure 6 below to predict the probability of selecting a man between a<br>
height of 72 and 76 in the given population (i.e the region highlighted in black in Figure 6).

````python
σ = 3       # std dev = 3
μ = 70      # mean = 70

rng = np.random.default_rng(100)                # instansiate seeded object of numpy generator class
array = rng.normal(μ, σ, 100000)                # generate normal array, mean of 70, std dev of 3

# calculate probability via cumulative desity function between range 72 to 76
probability = np.round(scipy.stats.norm.cdf(76, μ, σ)- scipy.stats.norm.cdf(72, μ, σ), 2)

def get_yfrequency(number):
    '''this returns the y co-ordinate of normal distribution with given x number'''
    return scipy.stats.norm.pdf(number , μ, σ);
````

````python
# plot hist, store count and bins in vars 
count, bins, _ = plt.hist(array, 100, density = True) 

# define probability density function, adopted from numpy official docs
f = 1 /(σ * np.sqrt(2 * np.pi)) * np.exp(-(bins - μ) ** 2 / (2 * σ ** 2))

# plot probability density line
plt.plot(bins, f, color = 'red', linewidth = 3);     

# format plot
plt.figtext(0.5, 0.0, 'Figure 6', wrap = True, horizontalalignment = 'center', fontsize = 15)
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Mens Height normal distribution model:\nThe probability of selecting a man between a \
height of 72 and 76 inches is {}'.format(probability))


# store current axis in variable and get upper limit of plot 
ax = plt.gca()                         
y_upper_lim = ax.get_ylim()[1]

# define array from 72 to 76 in 0.05 increments
shaded_range = np.arange(72, 76, 0.05)

# iterate range and shade area under curve
for num in shaded_range:
    ymax_num = get_yfrequency(num)/y_upper_lim
    plt.axvline(x = num, ymin = 0, ymax = ymax_num, color = 'black', linewidth = 5);
````

These plots are very interesting as they esentially combine boxplots and kernal density plots and compliment eachother very well.<br>
Lewinson (2019) details how, like the boxplot, the violin plot shows the median, 25th percentile, 75th percentile, upper / lower adjacent<br>
values and outliers. To add to this it shows the entire distribution flipped on the side of the altered boxplot. Lets explore the violin plot<br>
further:

- The interquartile range (IQR) is comprised of the Q1 (25th percentile), median (50th percentile) and Q3 (75th percentile). The<br>
percentile tells you that n % of values lie below the given percentile. For example looking at the upper quartile one would be able<br>
to deduce that 25% of the values lie above this marker.
- The min / max points indicate the lowest / highest values and can be derived via nth quartile +/- 1.5 IQR.
- In addition to seeing the data spread one can also see the entire distribution of the data via the kernal density estimator. This is<br>
extremely useful as in some cases it can be hard to visualise the distribution from a boxplot. For example looking at image 2 above<br>
it's very ovious that the data is normally distributed.
- The plot also yields information on outliers. The range from the min or max to their distribution tail resresents the distribution of<br>
outliers in the data.


````python
# code customised from:
# https://matplotlib.org/stable/gallery/statistics/customized_violin.html#sphx-glr-gallery-statistics-customized-violin-py

def format_violin_plot(data, x_labels):
    
    # calc the n-th percentile of the data along first axis
    quartile1, median, quartile3 = np.percentile(data, [25, 50, 75], axis = 1)

    # decorating the violin plot
    x_series = np.arange(1, len(median) + 1)                                               # define x series 
    plt.scatter(x_series, median, marker = 'o', color = 'w', s = 50, zorder = 3)           # plot median marker
    plt.vlines(x_series, quartile1, quartile3, color = 'black', linestyle = '-', lw = 10)  # plot interquartile range   
    xticks_list = np.insert(x_labels, 0, '', axis = 0)                                     # add space to start of x-labels list
    plt.xticks(np.arange(len(xticks_list)), xticks_list)                                   # pass xtick list (defined previously) 
    return;
````

### Scatter Plot
#### Plot Description
***

Scatter plots are very useful plots as they can show the association between two variables plotted agaisnt eachother. McDonald (2021) details<br>
how scatter plots work best when analyzing two continuous variables (floats) as plotting discrete variables results in a higher probability of<br>
data points overlapping and loss of information. Although scatter plots dont give quantitative data on the level of association between<br>
variables, one can quickly get visual information on the expression of quantitative change. Figure 8 below shows how visual a scatter plot <br>
can be in revealing different levels of correlation between two variables, correlation referring to the strength of association and the direction <br>of the
relationship [4].


````python
# plot description code
# generate 1x3 grid
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (15, 5))
plt.figtext(0.5, -0.05, 'Figure 8',wrap = True, horizontalalignment = 'center', fontsize = 15)

# generate two correlated normal arrays
mean = (0, 0)

# generate 3x correlated bivariate arrays by passing mean and correlation matrix
scatter_subplot1 = rng.multivariate_normal(mean, [[1, 0], [0, 1]], 1000)  
scatter_subplot2 = rng.multivariate_normal(mean, [[1, 0.95], [0.95, 1]], 1000)  
scatter_subplot3 = rng.multivariate_normal(mean, [[1, -0.95], [-0.95, 1]], 1000)  

ax1.scatter(scatter_subplot1[:, 0], scatter_subplot1[:, 1])
ax2.scatter(scatter_subplot2[:, 0], scatter_subplot2[:, 1])
ax3.scatter(scatter_subplot3[:, 0], scatter_subplot3[:, 1])                   

ax1.set_title('No Correlation')
ax2.set_title('Strong Positive Correlation')
ax3.set_title('Strong Negative Correlation');
````
